```python
import pandas as pd
data = {
    'Dataset': [
        'Original',
        'Noise Augmentation',
        'Pitch Augmentation',
        'Stretch Augmentation',
        'Mixed Augmentation'
    ],
    'Best Test Accuracy': [
        0.3333,
        0.4167,
        0.4444,
        0.3889,
        0.5278
    ]
}

df = pd.DataFrame(data)

print(df.to_markdown(index=True, tablefmt="grid"))
```

```python
import matplotlib.pyplot as plt


datasets = ["Original", "Noise Augmentation", "Pitch Augmentation", "Stretch Augmentation", "Mixed Augmentation"]
accuracies = [0.33, 0.42, 0.44, 0.39, 0.53]

color = "#5DADE2"

plt.figure(figsize=(10, 6))
bars = plt.bar(datasets, accuracies, color=color)

for bar in bars:
    bar.set_edgecolor('none')

plt.title("Comparison of Best Test Accuracy across Augmentations", fontsize=14)
plt.ylabel("Best Test Accuracy", fontsize=12)
plt.xlabel("Dataset", fontsize=12)
plt.xticks(rotation=20)

for i, acc in enumerate(accuracies):
    plt.text(i, acc + 0.01, f"{acc:.2f}", ha='center', fontsize=11)

plt.tight_layout()
plt.show()
```
